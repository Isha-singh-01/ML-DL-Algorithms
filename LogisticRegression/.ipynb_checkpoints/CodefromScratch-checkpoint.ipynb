This notebook implements logistic regression using 3 approaches:
1) Perceptron using step function
2) Perceptron using sigmoid function
3) Gradient descent with log loss function

```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
X,y = make_classification(n_samples=100,n_features=2,n_informative=1, n_redundant=0, n_classes=2, 
                          n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=20)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
## Using perceptron with step function
def perceptron_step(X,y):
    X = np.insert(X,0,1,axis=1) #Appending 1 to the array.
    w = np.ones(X.shape[1]) #We start with the initial values of features as 1.
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],w))
        w = w + lr*(y[j]-y_hat)*X[j]

    return w[0], w[1:]
```

```python
intercept_, coef_ = perceptron_step(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
X_input = np.linspace(-3,3,100)
Y_input = m*X_input + b
```

```python
figure = plt.figure(figsize=(8,4))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
plt.plot(X_input,Y_input,color='red',linewidth=3)
plt.ylim(-3,2)
```

```python
## perceptron using the sigmoid function
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def perceptron_sigmoid(X,y):
    X = np.insert(X,0,1,axis=1)
    w = np.ones(X.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = sigmoid(np.dot(X[j],w))
        w = w + lr*(y[j]-y_hat)*X[j]
    return w[0], w[1:]
```

```python
intercept_s, coef_s = perceptron_sigmoid(X,y)
```

```python
m = -(coef_s[0]/coef_s[1])
b = -(intercept_s/coef_s[1])
```

```python
X_input1 = np.linspace(-3,3,100)
y_input1 = m*X_input1 + b
```

```python
figure = plt.figure(figsize=(8,4))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
plt.plot(X_input,Y_input,color='red',linewidth=3)
plt.plot(X_input1,y_input1,color='orange',linewidth=3)
plt.ylim(-3,2)
```

Orange line is more balanced than the red one.

```python
## using scikit-learn
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X,y)
```

```python
m = -(lr.coef_[0][0]/lr.coef_[0][1])
b = -(lr.intercept_/lr.coef_[0][1])
```

```python
X_input2 = np.linspace(-3,3,100)
y_input2 = m*X_input2 + b
```

```python
figure = plt.figure(figsize=(8,4))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
plt.plot(X_input,Y_input,color='red',linewidth=3)
plt.plot(X_input1,y_input1,color='orange',linewidth=3)
plt.plot(X_input2,y_input2,color='black',linewidth=3)
plt.ylim(-3,2)
```

Black line represents the best fit line. We can see that using the sigmoid function, we reached close to the best fit line but it is still not the best. Hence, we study loss function and MLE.

```python
def gd(X,y):
    X = np.insert(X,0,1,axis=1)
    w = np.ones(X.shape[1])
    lr = 0.5

    for i in range(5000):
        y_hat = sigmoid(np.dot(X,w))
        w = w + lr*(np.dot((y-y_hat),X)/X.shape[0])
    return w[1:],w[0]
```

```python
coef_gd,intercept_gd = gd(X,y)
```

```python
m = -(coef_gd[0]/coef_gd[1])
b = -(intercept_gd/coef_gd[1])
```

```python
X_input3 = np.linspace(-3,3,100)
y_input3 = m*X_input3 + b
plt.figure(figsize=(10,6))
plt.plot(X_input,Y_input,color='red',linewidth=3)
plt.plot(X_input1,y_input1,color='orange',linewidth=3)
plt.plot(X_input2,y_input2,color='black',linewidth=3)
plt.plot(X_input3,y_input3,color='blue',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

Gradient descent fits much better

So, in this notebook we saw the improvement in building the best model for logistic regression.
